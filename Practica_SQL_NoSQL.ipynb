```python
import sqlite3
import pandas as pd

# Crear una conexión a la base de datos SQLite
conn = sqlite3.connect('ventas.db')

# Crear una tabla de ventas
query_create_table = '''
CREATE TABLE IF NOT EXISTS ventas (
    id_venta INTEGER PRIMARY KEY,
    fecha TEXT,
    producto TEXT,
    cantidad INTEGER,
    precio REAL
);
'''

# Ejecutar la creación de la tabla
conn.execute(query_create_table)
conn.commit()

# Simulación de datos de ventas
data_ventas = [
    ('2024-10-01', 'Laptop', 2, 1500.00),
    ('2024-10-02', 'Smartphone', 3, 800.00),
    ('2024-10-02', 'Tablet', 1, 600.00),
    ('2024-10-03', 'Audífonos', 5, 100.00),
]

# Insertar datos en la tabla
conn.executemany('INSERT INTO ventas (fecha, producto, cantidad, precio) VALUES (?, ?, ?, ?)', data_ventas)
conn.commit()

# Consultar la tabla
df_ventas = pd.read_sql_query('SELECT * FROM ventas', conn)
print(df_ventas)

# Cerrar la conexión
conn.close()
```

```python
def consolidar_datos():
    # Cargar datos de ventas
    conn = sqlite3.connect('ventas.db')
    df_ventas = pd.read_sql_query('SELECT * FROM ventas', conn)
    conn.close()

    # Cargar datos de reseñas
    df_reseñas = pd.DataFrame(list(collection.find()))

    # Consolidar ambos datasets
    df_consolidado = pd.merge(df_ventas, df_reseñas, on="producto", how="left")
    print("Datos consolidados:\n", df_consolidado)

    # Guardar en un nuevo archivo CSV (simulación de almacén de datos)
    df_consolidado.to_csv('consolidado_ventas_reseñas.csv', index=False)
    print("Consolidación completada y almacenada.")

# Ejecutar consolidación manualmente (puede ser programado)
consolidar_datos()
```

```python
import plotly.express as px
import pandas as pd
from pymongo import MongoClient

# Conexión a MongoDB Atlas (Reemplaza con tus credenciales)
client = MongoClient('mongodb+srv://databaseuser:password@clustertalentotech.rxc9u.mongodb.net/?retryWrites=true&w=majority&appName=ClusterTalentoTech')
db = client["reseñas_db"]
collection = db["reseñas"]

# Consultar datos de la colección
reseñas = collection.find()

# Convertir los datos a DataFrame
df_reseñas = pd.DataFrame(list(reseñas))

# Limpiar el DataFrame de posibles valores nulos
df_reseñas.fillna('', inplace=True)

### Gráfica 1: Gráfica de Barras (Distribución de Calificaciones) ###

# Crear una gráfica de barras con calificaciones segmentadas por producto
fig_barras = px.bar(df_reseñas, x='producto', y='calificación', color='calificación',
                    title="Distribución de Calificaciones por Producto",
                    labels={'calificación': 'Calificación', 'producto': 'Producto'},
                    text_auto=True)

# Mostrar la gráfica de barras
fig_barras.show()


### Gráfica 2: Gráfica de Pastel (Porcentaje de Calificaciones) ###

# Crear una gráfica de pastel segmentada por nombre de producto
fig_pie = px.pie(df_reseñas, names='producto', values='calificación',
                 title="Porcentaje de Calificaciones por Producto",
                 labels={'producto': 'Producto', 'calificación': 'Calificación'})

# Mostrar la gráfica de pastel
fig_pie.show()
```
