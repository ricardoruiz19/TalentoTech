```python
from datetime import datetime
# lee el archivo excel Datos+Empleados.xlsx
df = pd.read_excel('Datos+Empleados.xlsx')
# calcula la edad de los empleados a partir de la fecha de nacimiento
def calculate_age(born):
  today = datetime.today()
  return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

df['Edad'] = df['Nacimiento'].apply(calculate_age)
df.head()
```

```python
# calcula la edad promedio de los empleados
edad_promedio = int(df['Edad'].mean())
print("La edad promedio de los empleados es:", edad_promedio)
```

```python
# crea un total de los empleados por estado
empleados_agrupados = df.groupby('Estado')['ID Empleado'].count()
print(empleados_agrupados)
```

```python
# crea una tabla visual con nombre empleado, departamento, posición, estado, genero
import plotly.graph_objects as go

fig = go.Figure(data=[go.Table(
    header=dict(values=['Nombre Empleado', 'Departamento', 'Posición', 'Estado', 'Género'],
                fill_color='paleturquoise',
                align='left'),
    cells=dict(values=[df['Nombre Empleado'], df['Departamento'], df['Posición'], df['Estado'], df['Género']],
               fill_color='lavender',
               align='left'))
])

fig.show()
```

```python
# en un grafico de barras muestra la cantidad de empleados por departamento, organizalos de mayor a menor
import plotly.express as px

# Cuenta los empleados por departamento y ordena de mayor a menor
empleados_por_depto = df['Departamento'].value_counts()

# Crea el gráfico de barras
fig = px.bar(x=empleados_por_depto.index, y=empleados_por_depto.values,
             labels={'x':'Departamento', 'y':'Número de Empleados'},
             title='Cantidad de Empleados por Departamento')

fig.show()
```
